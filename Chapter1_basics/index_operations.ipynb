# INDEX OPERATIONS
- Input: `df.index`. Output: RangeIndex(start=, stop=, step=1)
- `inplace=True` method tells the function to modify the original DataFrame directly rather than creating a new one. 


```python
import pandas as pd
import csv

# Importing CSV dataset
df = pd.read_csv("studentsperformance.csv")
```

```python
#Printing after df.set_index to verify Index
df.index
```

- You can set any column as an index, instead of index = 0,1,2.....
- Syntax: `df.set_index("column name", inplace=True)`.
- So after updating DataFrame, if u print df.index, `index` will give be substituted.

```python
df.set_index('parental_level_of_education', inplace=True)
df.head()
```

```python
#Printing after df.set_index to verify Index
df.index
```

##### Reseting Index
- `df.reset_index(inplace=True)`

```python
df.reset_index(inplace=True)
df
```
