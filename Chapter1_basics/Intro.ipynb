# PANDAS LIBRARY
**Definition:**
- Pandas is a data manipulation package in Python for tabular data.
- Data in the form of rows and columns, also known as DataFrames. 
- Functionality includes: 

            1. Data transformations; Like sorting rows and taking subsets.
            2. calculating summary statistics such as the mean, reshaping DataFrames, and joining DataFrames together.
- Pydata Ecosystem.

**pandas UseCase**
- Most Imp in data analysis workflow.
- Import files and datasets etc.
- Clean Datasets.
- Structure datasets into a suitable format for analysis.
- Visualize datasets and uncover insights.
- Time series analysis and analyzing text data.

**Benefits**
- pandas is less verbose, requiring fewer lines of code to get the desired output. 
- intuitive data representation that facilitates easier data understanding and analysis.
- Supports an extensive set of operations from exploratory data analysis, dealing with missing values, calculating statistics, visualizing univariate and bivariate data, and much more.
- pandas handles large data sets with ease. It offers speed and efficiency.

**Basic Data Structure**
- **Series**:A Series is essentially a column.

- **DataFrame**:  A multi-dimensional table made up of a collection of Series.

**The Golden Rule**
- If a column name has spaces, brackets, parentheses, %, or symbols  ALWAYS use square brackets.
- Rename messy columns early for smooth analysis
 Correct Way to Fix Your Code:

1.  df.gender(f/m).mean()  fails for special characters.
2.  df['gender(f/m)'].mean()  always works

## .shape()
- Rows and Columns of a DataFrame can be identified using the .shape attribute of the DataFrame. 
- It returns a tuple (row, column) and can be indexed to get only rows, and only columns count as output.
- Syntax: 

        1. `df.shape`: Returns number of rows and columns.
        2. `df.shape[0]: Returns number of rows only.
        3. `df.shape[1]: Returns number of columns only.


```python
df.shape
print("Number of Rows in dataset:",df.shape[0])
print("Number of Columns in dataset:",df.shape[1])
```
