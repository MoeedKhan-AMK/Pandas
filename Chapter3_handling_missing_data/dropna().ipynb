### dropna() ->Delete Rows or Columns with Missing Data
- A core method used to remove missing values (NaN, None, or NaT) from a DataFrame. 
- Allows you to clean your dataset by dropping entire rows or columns based on different criteria, which is a crucial step in data preparation and analysis. 
- The dropna() method offers several parameters for fine-grained control over which data is removed.
- It will simply delete entire row or column, with the help of given parameter.
- `df.dropna()`
- Useful when:
    1. Missing data is large
    2. Row is meaningless without the missing field
    3. Data quality needs to be strict

```python
df.dropna()
```

#### PARAMETERS
- `how`: Determines the condition for dropping. Following two are condition for `how` parameter:
    1. 'any' (default): Drop the row/column if even one missing value is present.
    2. 'all': Drop the row/column only if all values are missing.
- `thresh`: Requires a minimum number of non-NA values to keep the row/column. This parameter cannot be used in combination with `how`!.

    Simply `thresh` is a threshold that if you set it to `1`
- `subset`: Restricts the search for missing values to specific columns.

```python
df.dropna(how='all')        #will only drop if all values are missing
```

```python
df.dropna(how='any')        #will drop even if single value is missing
```

```python
df.dropna(thresh=1)
```
